### Sample program - Google's PageRank Algorithm

This program is from a project I did for my Scientific Programming class where I explored how the early version of Google's PageRank algorithm works. PageRank works by counting the number and the quality of links to a page determin a rough estimate of how important the website is. The underlying assumption is that important websites are likely to recieve links from other important websites.

The following program demonstrates one way to calculate PageRank called power method. It takes in the 2D which represent the weights of links between two websites and numbers of iterations you want to perform. The function returns a vector that represent the importance of each website. 

The table below shows how the importance of the websites change with each iteration.

```python
%%HTML
<style type='text/css'>
    table.dataframe td, table.dataframe th {
        border-style: solid;
        border-color: #84a9ac;
        border-width: thin;
    }
</style>
```

```python
import numpy as np
import pandas as pd
import math
import scipy.linalg
import pprint
import random
pd.set_option('max_colwidth', 400)

L = np.array([
    [0, 0, 1/2, 1/2],
    [1, 0, 0, 1/2],
    [0, 1/2, 0, 0],
    [0, 1/2, 1/2, 0]
]) # expected rank = [.217, .349, .174, .261]

# normalize so all elements add to 1
def normalize(v):
    return v / np.sum(v)

def eigenvalue(A, v):
    Av = A.dot(v)
    return v.dot(Av)

def power_method_with_iter(A, iters):
    n, d = A.shape
    vectors = []
    
    v = np.ones(n) / n
    ev = eigenvalue(A, v)
    
    iterations = 0
    vectors.append({'vec': v})
    
    for i in range(iters):
        Av = A.dot(v)
        v_new = Av / np.linalg.norm(Av)
        ev_new = eigenvalue(A, v_new)
        
        vectors.append({'vec': normalize(v_new)})
        
        ev = ev_new
        v = v_new

    return vectors

vec = power_method_with_iter(L, 30)
df = pd.DataFrame(vec)
df
```
